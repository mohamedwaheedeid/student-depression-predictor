```python
scale = StandardScaler()

# Scale the numerical columns from the original X_train and X_test
X_train_num_scaled = scale.fit_transform(X_train[num_cols])
X_test_num_scaled = scale.transform(X_test[num_cols])

# Convert one-hot encoded sparse matrices to dense arrays
X_train_cat_dense = X_train_ohe.toarray()
X_test_cat_dense = X_test_ohe.toarray()

# Concatenate scaled numerical features with dense one-hot encoded categorical features
X_train_scaled = np.hstack((X_train_num_scaled, X_train_cat_dense))
X_test_scaled = np.hstack((X_test_num_scaled, X_test_cat_dense))
# convering numpy array to pandas dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_cols + list(ohe.get_feature_names_out(cat_cols)))
```

```python
import optuna
```

```python
study = optuna.create_study(direction="maximize")
```

```python
# getting cross_val_score
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
```

```python
def objective(trial):

    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 500),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "tree_method": "hist",
        "device": "cuda",
    }

    model = XGBClassifier(**params, enable_categorical=True)

    model.fit(X_train_pca, y_train)

    return model.score(X_test_pca, y_test)
```

```python
study.optimize(objective, n_trials=50)
```

```python
best_params = study.best_params
best_params
```

```python
final_model = XGBClassifier(
    **best_params,
    tree_method="hist",
    device="cuda",
    random_state=42
)

final_model.fit(X_train_pca, y_train)
```

```python
#show predictions as scores
ypred = final_model.predict(X_test_pca)
accuracy = accuracy_score(y_test, ypred)
print(f"Accuracy of the final model: {accuracy:.4f}")
```
